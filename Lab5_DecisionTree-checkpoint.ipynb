```python
import pandas as pd
import matplotlib as plt
```

```python
df=pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
#2.	Split the dataset into independent variables (features) and the dependent variable (target).
y=df['Heart Disease']
x=df.drop(['Heart Disease'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree

model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)

# F1-score
f1 = f1_score(y_test, y_pred)
print("F1-score:", f1)
```

```python
ind_feature_name = x.columns.tolist()
cls_name = [str(cls) for cls in y.unique().tolist()]
plot_tree(model, filled=True, max_depth=None, node_ids=True, feature_names=ind_feature_name, class_names=cls_name)
```
