```python

import pandas as pd
import matplotlib as plt

```

```python
df= pd.read_csv('Lab_1.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
features_with_na= [features for features in df.columns if df[features].isnull().sum()>1]
features_with_na
```

```python
#filling features with na with the mean value 
for feature in features_with_na:
    df[feature].fillna(df[feature].mean, inplace=True)
```

```python
numerical_summary = df.describe()
print("Statistical summary of numerical variables:")
print(numerical_summary)
```

```python
#Visualize the distribution of numerical variables using histograms
import matplotlib.pyplot as plt

# Plot histograms for numerical variables
df.hist(figsize=(10, 8))
plt.suptitle('Histograms of Numerical Variables', fontsize=16)
plt.show()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Study Hours'] = pd.to_numeric(df['Study Hours'], errors='coerce')

# Plot scatter plot for numerical variables
plt.scatter(df['Age'], df['Study Hours'])
plt.title('Scatter Plot of Age vs Study Hours')
plt.xlabel('Age')
plt.ylabel('Study Hours')
plt.show()
```

```python
# Plot bar plots for categorical variables
categorical_columns = ['Gender']
for col in categorical_columns:
    plt.figure(figsize=(8, 6))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
